```python
import os
import random
import shutil
```

```python
def train_val_split(split = 0.8):
    image_classes = os.listdir("JPEGImages")
    for cl in image_classes:
        cl_images = os.listdir("JPEGImages/"+str(cl))
        random.Random(95).shuffle(cl_images)
#         print(cl_images)
        train_images_indices = int(len(cl_images)*split)
        train_images = cl_images[:train_images_indices]
        val_images = cl_images[train_images_indices:]
        
        print(train_images)
        print(val_images)
        
#         train_destination = "Train_"+str(split)+"/"+str(cl)
        
        for image in train_images:
            if not os.path.exists("Train_"+str(split)+"/"+str(cl)):
                os.mkdir("Train_"+str(split)+"/"+str(cl))
            train_source = "JPEGImages/"+str(cl)+"/"+image
            train_destination = "Train_"+str(split)+"/"+str(cl)
#             print(train_source)
#             print(train_destination)
            shutil.copy(train_source,train_destination)
        
        for image in val_images:
            if not os.path.exists("Validation_"+str(split)+"/"+str(cl)):
                os.mkdir("Validation_"+str(split)+"/"+str(cl))
            val_source = "JPEGImages/"+str(cl)+"/"+image
            val_destination = "Validation_"+str(split)+"/"+str(cl)
#             print(val_source)
#             print(val_destination)
            shutil.copy(val_source,val_destination)
            

train_val_split(split = 0.8) 

```

```python
def count_train_test_images(train_dir = "Train_0.8", val_dir = "Validation_0.8"):
    image_classes = os.listdir(train_dir)
    train_image_len = 0
    val_image_len = 0
    for cl in image_classes:
        train_images = os.listdir(train_dir+"/"+str(cl))
        train_image_len += len(train_images)
        val_images = os.listdir(val_dir+"/"+str(cl))
        val_image_len += len(val_images)
    print(train_image_len)
    print(val_image_len)
        
#         print(train_images)
#         print(val_images)


count_train_test_images(train_dir = "Train_0.8", val_dir = "Validation_0.8")
```
